 ## 1. Characterize the Airbnb locations distribution in the city

### 1.1. Prepare data

The data set all_date_listings has all the listings availables for 4 dates:
    - December 2018
    - April 2020
    - December 2020
    - April 2021
    
For every date there is a entry for each listing/location. It has data about the type of accomodation, the location in the city, the district and neigbourhood it belongs, the total number of reviews accumulated till that date, the average review received, and the average price,...

```python
all_dates_listings = pd.read_csv("../data/clean_data/all_dates_listings.csv")
```

#### 1.1.1. Finish to clean data set

```python
all_dates_listings.columns
```

```python
#we will keep those columns
listings= all_dates_listings[['id','neighbourhood','latitude', 'longitude', 'room_type', 'price', 'minimum_nights',
                   'number_of_reviews','reviews_per_month','calculated_host_listings_count', 
                   'availability_365', 'Date','District_code','District']]
```

```python
#The missing values are because they did not have reviews, therefore must be better to just fill those
#values with 0
listings['reviews_per_month'].fillna(0,inplace=True)
```

#### 1.1.2. Add information about safety and green areas in the data set

```python
district_descriptors = pd.read_csv("../data/clean_data/district_descriptors.csv")
district_descriptors = district_descriptors.iloc[:,1:]
district_descriptors 
```

```python
district_descriptors = district_descriptors.groupby(["District_name","District_code"]).agg(
    {"average_number_of_crimes":"sum","green_areas_m2":"mean","num_green_areas":"mean"}).reset_index()

district_descriptors = district_descriptors.rename(columns={"District_name":"District"})
listings_with_descriptors = listings.merge(district_descriptors, left_on=['District_code',"District"],right_on=['District_code',"District"],how="left")
listings_with_descriptors
```

```python
#how many bookings/reviews x day along time
review_avgprice_xday = calendar_reviews.groupby("date").agg({"reviewed":
                                      "sum","price":"median"}).rename(columns= {"reviewed":
                                                                                "reviews","price":"median_price"}
                                                                     ).reset_index()
```

```python
#https://www.dataquest.io/blog/tutorial-time-series-analysis-with-pandas/
```

```python
review_avgprice_xday = review_avgprice_xday.set_index("date")
review_avgprice_xday.head()
```

```python
#I compute the average mean x month
review_avgprice_xmonth= review_avgprice_xday.resample('M').mean()
```

```python
sns.color_palette("Set2")
current_palette = sns.color_palette()
```

```python
#add vertical lines

# ax = review_avgprice_xday.loc['2018-12-06':'2021-12','median_price'].plot( linestyle='-')
# ax.set_ylabel('median price per day(euros)');

# Start and end of the date range to extract
start, end = '2018-12-06', '2021-12'
# Plot daily and weekly resampled time series together
fig, ax = plt.subplots(figsize = (20,5))
ax.plot(review_avgprice_xday.loc[start:end, 'median_price'],
marker='.', linestyle='-', linewidth=0.5, label='Median Daily price')
ax.plot(review_avgprice_xmonth.loc[start:end, 'median_price'],
marker='o', markersize=8, linestyle='-', label='Monthly Median price')
ax.set_ylabel("median price per day(euros)")
ax.legend('')
plt.axvline(pd.to_datetime('2020-03-15'), color='r', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2020-12-15'), color='r', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2020-07-01'), color='b', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2021-04-20'), color='b', linestyle='--', lw=2)
plt.axvline(pd.to_datetime('2020-08-18'), color='r', linestyle='--', lw=2)
plt.savefig("../figures/price_over_time.png",dpi=600)
```
